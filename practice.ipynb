```python
# Bubble sort
for i in range(len(l)):
    for j in range(len(l)-1):
        if l[j]>l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]
print(l)
```

```python
def duplicates(l:list)->set:
    dup=set()
    st1=set()
    for i in l:
        if i in st1:
            dup.add(i)
        else :
            st1.add(i)
    return dup
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(3):

    # Roll two dice 1000 times
    dice1 = np.random.randint(1, 7, 1000)
    dice2 = np.random.randint(1, 7, 1000)
    sums = dice1 + dice2

    # Plot
    plt.hist(sums, bins=11, edgecolor='black')
    plt.title("Distribution of Dice Sums (1000 Rolls)")
    plt.xlabel("Sum")
    plt.ylabel("Frequency")
    plt.show()
```
